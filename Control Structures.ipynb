```python
# Simple if statement
age = 25
if age >= 18:
    print("Adult")
```

```python
# if-else statement
temperature = 15
if temperature > 20:
    print("Warm day")
else:
    print("Cool day")
```

```python
# if-elif-else chain
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f"Grade: {grade}")
```

```python
# Syntax: value_if_true if condition else value_if_false
age = 20
status = "adult" if age >= 18 else "minor"

# Nested ternary (use sparingly for readability)
number = 0
result = "positive" if number > 0 else "negative" if number < 0 else "zero"

# Practical example
def get_max(a, b):
    return a if a > b else b
```

```python
# Multiple conditions with membership testing
username = "admin"
if username in ["admin", "root", "superuser"]:
    print("Administrator detected")

# Chained comparisons
value = 50
if 0 <= value <= 100:
    print("Value in valid range")

# Checking for None, empty collections
data = []
if data:  # Pythonic way to check if list is non-empty
    print("Data available")
else:
    print("No data")

# Walrus operator (Python 3.8+) - assignment within condition
if (n := len(data)) > 10:
    print(f"Large dataset with {n} items")

# Match-case statement (Python 3.10+)
def http_status(status_code):
    match status_code:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500 | 502 | 503:  # Multiple patterns
            return "Server Error"
        case _ if status_code >= 400:  # Guard clause
            return "Client Error"
        case _:
            return "Unknown"
```

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Iterating over a string
for char in "Python":
    print(char)

# Using range()
for i in range(5):  # 0, 1, 2, 3, 4
    print(i)

for i in range(2, 10):  # 2 through 9
    print(i)

for i in range(0, 10, 2):  # 0, 2, 4, 6, 8 (step of 2)
    print(i)

# Reverse iteration
for i in range(10, 0, -1):
    print(i)
```

```python
# enumerate() - get index and value
students = ["Alice", "Bob", "Charlie"]
for index, student in enumerate(students):
    print(f"{index}: {student}")

# Start enumerate at different index
for index, student in enumerate(students, start=1):
    print(f"Student {index}: {student}")

# zip() - iterate over multiple sequences
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
cities = ["NYC", "LA", "Chicago"]

for name, age, city in zip(names, ages, cities):
    print(f"{name}, {age}, lives in {city}")

# zip_longest for unequal sequences (from itertools)
from itertools import zip_longest
list1 = [1, 2, 3]
list2 = ['a', 'b']
for num, letter in zip_longest(list1, list2, fillvalue='N/A'):
    print(num, letter)
```

```python
student_grades = {"Alice": 90, "Bob": 85, "Charlie": 92}

# Iterate over keys (default)
for name in student_grades:
    print(name)

# Explicit key iteration
for name in student_grades.keys():
    print(name)

# Iterate over values
for grade in student_grades.values():
    print(grade)

# Iterate over key-value pairs
for name, grade in student_grades.items():
    print(f"{name}: {grade}")
```

```python
# Basic list comprehension
squares = [x**2 for x in range(10)]

# With conditional
even_squares = [x**2 for x in range(10) if x % 2 == 0]

# With if-else (must come before 'for')
result = [x if x % 2 == 0 else -x for x in range(10)]

# Nested comprehension
matrix = [[i*j for j in range(3)] for i in range(3)]

# Dictionary comprehension
word_lengths = {word: len(word) for word in ["apple", "banana", "cherry"]}

# Set comprehension
unique_lengths = {len(word) for word in ["apple", "app", "banana"]}

# Generator expression (memory efficient)
sum_of_squares = sum(x**2 for x in range(1000000))
```

```python
# Simple counting
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Exit loop early
for i in range(100):
    if i == 5:
        break  # Exit loop when i is 5
    print(i)

# Finding first match
numbers = [3, 7, 2, 9, 12, 5]
target = 9
for num in numbers:
    if num == target:
        print(f"Found {target}!")
        break
else:  # else clause executes if loop completes without break
    print(f"{target} not found")

# Breaking from nested loops
found = False
for i in range(5):
    for j in range(5):
        if i * j == 12:
            print(f"Found at ({i}, {j})")
            found = True
            break
    if found:
        break
```

```python
# Nested if statements
def classify_temperature(temp, humidity):
    if temp > 30:
        if humidity > 70:
            return "Hot and humid"
        else:
            return "Hot and dry"
    elif temp > 20:
        if humidity > 70:
            return "Warm and humid"
        else:
            return "Pleasant"
    else:
        return "Cool"

# Refactored with early returns (preferred)
def classify_temperature_better(temp, humidity):
    if temp > 30 and humidity > 70:
        return "Hot and humid"
    if temp > 30:
        return "Hot and dry"
    if temp > 20 and humidity > 70:
        return "Warm and humid"
    if temp > 20:
        return "Pleasant"
    return "Cool"
```

```python
# Continue on error
data = [10, 0, 5, 'invalid', 20]
for item in data:
    try:
        result = 100 / float(item)
        print(f"Result: {result}")
    except (ValueError, ZeroDivisionError) as e:
        print(f"Error processing {item}: {e}")
        continue
```
